```python
x = np.arange(0, 10)
y = x*x

plt.title("Line Plot")
plt.xlabel('X-axis', color='magenta')
plt.ylabel('Y-axis', color='magenta')
plt.plot(x, y, 'co--', linewidth=1.5, markersize=5)
```

```python
x = np.arange(1, 10)
y = 3 * x + 10
plt.title('2d-Diagram', color='blue')
plt.xlabel('X-axis', color='blue')
plt.ylabel('Y-axis', color='blue')
plt.plot(x, y, 'r*--', linewidth=2, markersize=8)
```

```python
x = [1,2,3,4,5,6,7]
y = [160,150,140,145,175,165,180]

plt.plot(x,y, 'mx-', markersize=7)
plt.title("Total Sales of Companies in a Week of Each day")
plt.xlabel('Days of a d week', color='green')
plt.ylabel('Sales', color='green')
plt.show()
```
