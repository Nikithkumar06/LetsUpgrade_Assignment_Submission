```python
import numpy as np
import pandas as pd
import seaborn as sns

import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#QUESTION 1:- plot a line plot between x and y:
x=np.arange(0,10)
y=x*x

plt.title("Graph Between X and Y")
plt.xlabel('X-axis', color='black')
plt.ylabel('Y-axis', color='black')
plt.plot(x, y, 'co--', linewidth=1.5, markersize=5)
```

```python
#QUESTION 2:- plot a line plot between a and b:
a=np.arange(40,50)
b=np.arange(50,60)
plt.plot(a,b,"mo--",linewidth=2, markersize=5)
plt.title("Line Plot between A and B")
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
```

```python
#QUESTION-3:-Plot a line plot showing the sales trend of company 1 and 2:

days = [1,2,3,4,5,6,7] # Days of the week
sales_1 = [160,150,140,145,175,165,180] # Company 1 sales 
sales_2 = [70,90,160,150,140,145,175]  # Company 2 sales 
plt.title("Distribution of Sales Trends of Company 1 & 2")
plt.plot(days, sales_1,'bx-',color='c' , label = 'Company 1 Sales')
plt.plot(days, sales_2,'bx-',color='b' , label = 'Company 2 Sales')
plt.xlabel("Days of the week")
plt.ylabel("Company Sales")
plt.legend(loc='lower right')
plt.show()
```

```python
#Question 4 :Create a 3 by 3 subplots:

x = [1,2,3,4]
y1 = [4,3,2,1]
y2 = [10,20,30,40]
y3 = [40,30,20,10]
y4 = [1,2,1,2]
y5 = [40,70,90,70]

plt.suptitle('Subplots:',fontsize=17)

plt.subplot(3,3,1)
plt.plot(x,y1,'c-')
plt.xlabel('x')
plt.ylabel('y1')

plt.subplot(3,3,2)
plt.plot(x,y2,'r*-')
plt.xlabel('x')
plt.ylabel('y2')

plt.subplot(3,3,3)
plt.plot(x,y3,'mx--')
plt.xlabel('x')
plt.ylabel('y3')

plt.subplot(3,3,4)
plt.plot(x,y4,'go--')
plt.xlabel('x')
plt.ylabel('y4')

plt.subplot(3,3,5)
plt.plot(x,y5,'y*--')
plt.xlabel('x')
plt.ylabel('y5')
plt.show()
```
