CATEGORICAL DATA PLOTTING

boxplot(), violinplot(), etc..,

```python
tips.head(5)
```

```python
titanic = sns.load_dataset('titanic')
titanic.head(5)
```

```python
sns.catplot(x = 'day', y = 'total_bill', data = tips)
```

```python
exercise = sns.load_dataset("exercise")
exercise.head(5)
```

```python
sns.catplot(x = 'time', y = 'pulse', hue = 'kind', data = exercise)
```

```python
sns.catplot(x = 'day', y = 'tip', kind = 'box', hue = 'sex', data = tips)
```

```python
diamonds = sns.load_dataset('diamonds')
diamonds.head()
```

```python
sns.catplot(x = 'color', y = 'price', kind = 'boxen', data = diamonds.sort_values('color'))
```

```python
sns.catplot(x = 'day', y = 'total_bill', kind = 'boxen', data = tips, dodge = False)
```

```python
sns.catplot(x = 'total_bill', y = 'day', hue = 'sex', kind = 'violin', data = tips)
```

```python
sns.catplot(x = 'deck', kind = 'count', palette = 'ch:0.95', data = titanic, hue = 'class')
```

```python
g = sns.catplot(x = 'day', y = 'total_bill', kind = 'violin',inner = None, data = tips)
sns.swarmplot(x = 'day', y = 'total_bill', color = 'k', size = 3, data = tips, ax = g.ax)
```

```python
sns.catplot(x = 'day', y = 'total_bill', hue = 'sex', kind = 'violin', data = tips, split = True)
```
